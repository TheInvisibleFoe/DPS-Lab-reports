```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use("seaborn-v0_8")
```

```python
counts = pd.read_excel("MCA.xlsx",header=0,nrows=1026).dropna()
counts
```

```python
Co_Counts = np.array(list(counts['Cobalt 600s']))
Cs_Counts = np.array(list(counts['Cesium 600s']))
channels = np.arange(0,1024,1)
```

```python
fig,axes = plt.subplots(1,1,figsize=(16,9))
axes.set_title("Cobalt counts fo 10 mins")
axes.set_xlabel("Channels")
axes.set_ylabel("Counts")
axes.plot(channels, Co_Counts)
```

```python
fig,axes = plt.subplots(1,1,figsize=(16,9))
axes.set_title(r"$_{55}Cs^{137}$ counts for 10 mins")
axes.set_xlabel("Channels")
axes.set_ylabel("Counts")
axes.plot(channels[:401], Cs_Counts[:401])
axes.annotate("Photopeak",(225,70000))
```
